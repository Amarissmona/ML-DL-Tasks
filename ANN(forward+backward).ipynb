```python
import numpy as np

#XOR DATASET
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[0]])
```

initialize *weights*

```python
np.random.seed(1)
W1 = np.random.randn(2,3)
W2 = np.random.randn(3,1)
#Input layer: 2 neurons
#Hidden Layer : 3 neurons
#Output Layer : 1 neuron
#Weight matrices :W1 (Input → Hidden)shape = (2 × 3)
#W2 (Hidden → Output) shape = (3 × 1)
```

*Activation*

```python
def sigmoid(X):
  return 1/(1+np.exp(-X))

def sigmoid_derivative(X):
  return X*(1-X)
```

```python
#Used in backpropagation to calculate gradient. It tells how much weights should change.
# training
for i in range(10000): #The network trains 10,000 times (epochs) to improve accuracy.

    # -------- Feedforward --------
    hidden = sigmoid(np.dot(X, W1))#Input → Hidden Layer atrix multiplication:
                                    #(4×2) dot (2×3) = (4×3)Then apply sigmoid activation.


    output = sigmoid(np.dot(hidden, W2)) #Hidden → Output Matrix multiplication:

                                        #(4×3) dot (3×1) = (4×1) Now we get predicte
    error = Y-output
#print(error)
    d_output = error * sigmoid_derivative(output)
#This calculates gradient for output layer.How much W2 should change
    error_hidden = d_output.dot(W2.T)
#popagates error back to hidden layer
    d_hidden = error_hidden * sigmoid_derivative(hidden)
#computes gradient for hidden laye neurons.
#update weights
    W2 += hidden.T.dot(d_output)
    W1 += X.T.dot(d_hidden)

print("Predicate Output:") #which is close to:[0,1,1,0]
print(output)
```
